# A* Pathfinding Demo

This notebook demonstrates the upgraded reusable A* module from this repository.

- Uses `astar_shortest_path` from `astar/algorithm.py`
- Runs the same sample graph as `data/sample_graph.json`
- Prints shortest path and total cost

```
from astar import astar_shortest_path

# same problem used in data/sample_graph.json
graph = {
    "S": {"A": 1, "B": 4},
    "A": {"B": 2, "C": 5, "D": 12},
    "B": {"C": 2},
    "C": {"D": 3},
    "D": {}
}
heuristic = {
    "S": 7,
    "A": 6,
    "B": 2,
    "C": 1,
    "D": 0
}

path, cost = astar_shortest_path(graph, heuristic, start="S", goal="D")
print("Path:", " -> ".join(path))
print("Cost:", cost)
```
